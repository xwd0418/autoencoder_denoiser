```python
real_img_loader = get_real_img_dataset(config)
raw_image_collection_real_data = []
noise_image_collection_real_data = []
for iter, data in enumerate(tqdm(real_img_loader)):
    noise, raw =  data
    raw_image_collection_real_data.append(np.array(raw[raw!=0].cpu()).flatten())
    noise_image_collection_real_data.append(np.array(noise[noise!=0].cpu()).flatten())
```

```python
plt.rcParams["figure.figsize"] = (20,10)
figure, plots = plt.subplots(ncols=2, nrows=1)
plots[0].imshow(noise[0,0],cmap=custom_HSQC_cmap, vmax=1, vmin=-1)
plots[1].imshow(raw[0,0],cmap=custom_HSQC_cmap, vmax=1, vmin=-1)
```

```python
import matplotlib.pyplot as plt
stacked_raw_img_real_data = np.concatenate(raw_image_collection_real_data)
stacked_noise_img_real_data = np.concatenate(noise_image_collection_real_data)
plt.rcParams["figure.figsize"] = (12,7)

plt.hist(stacked_raw_img_real_data)
plt.title("reference pixel distribution of real images", fontsize=20)
plt.figure()
plt.hist(stacked_noise_img_real_data)
plt.title("noisy pixel distribution of real images", fontsize=20)
plt.figure()

```
