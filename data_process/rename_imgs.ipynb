```python

import json
import os, re
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import json
import cv2
import shutil


img_dir = "/root/autoencoder_denoiser/dataset/real_noise_greyscale"
new_dir = "/root/autoencoder_denoiser/dataset/real_noise_renamed/"

# pattern = re.compile(r'np(7|8)()*_si_00()*页面*_(/d)*.npy')
pattern = re.compile(r'()*')
for img_folder in glob(img_dir+"/*/"):
    for img_path in os.listdir(img_folder):
        code = re.search("np(.+?)_si(.+?)", img_path).group(1)
        try:
            page = re.search("(.+?)_页面_(.+?).npy", img_path).group(2)
            shutil.copyfile(img_folder+img_path, (new_dir+code+"_"+page+".npy"))
        except:
            page = None
            shutil.copyfile(img_folder+img_path, (new_dir+code+".npy"))

```
