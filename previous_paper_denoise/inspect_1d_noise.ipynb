```python
import numpy as np
import os
from matplotlib import pyplot as plt 

```

```python
data_dir  = "training"
training_noise = [ os.path.join("gen_dataset", "1d", data_dir, f)\
    for f in os.listdir(os.path.join("gen_dataset", "1d", data_dir))]

data_dir  = "training_good"
training_ref = [ os.path.join("gen_dataset", "1d", data_dir, f)\
    for f in os.listdir(os.path.join("gen_dataset", "1d", data_dir))]

training_noise = sorted(training_noise)
training_ref = sorted(training_ref)
```

```python
for i in  range (len(training_noise)):
    if i == 0:
        for j in range(4):
            y_noise = np.loadtxt(training_noise[i])[j]
            y_ref = np.loadtxt(training_ref[i])
            # x = None
            plt.title("noise") 
            plt.plot(y_noise) 
            plt.show()
        
        plt.title("ref") 
        plt.plot(y_ref) 
        plt.show()
        
        break
```
